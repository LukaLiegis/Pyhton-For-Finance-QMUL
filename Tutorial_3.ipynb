```python
# Import various libraries from findatapy for downloading market data
from findatapy.market import Market, MarketDataRequest

# Create market object
market = Market()
```

```python
# Create a MarketDataRequest
# Hint: data_source parameter is 'yahoo'
# Ticker for Apple is aapl and for Tesla is tsla
# fields and vendor_fields should be adj close
md_request = MarketDataRequest(
    start_date="01 Jan 2005",
    finish_date="01 Jan 2022",
    tickers=["Apple", "Tesla"],
    vendor_tickers=["aapl", "tsla"],
    fields=["adj close"],
    vendor_fields=["adj close"],
    data_source="yahoo"
)

df_yahoo_stocks = market.fetch_market(md_request)
```

```python
df_yahoo_stocks
```

```python
# Resample data so it is end of month
# Hint: use pandas' resample function with last
df_monthly = df_yahoo_stocks.resample("M").last()
```

```python
# Calculate monthly returns
df_monthly_return = df_monthly / df_monthly.shift(1) - 1
```

```python
# Select only 2019 returns
# Hint: can use .index.year property on the DataFrame's index
df_2019_rets = df_monthly_return[df_monthly_return.index.year == 2019]
```

```python
# Plot 2019 returns as bar chart
df_2019_rets.plot(kind="bar")
```

```python
# Do the same thing for annual data for the full sample
# First resample the daily data into annual
# Hint: use 'Y' as resample parameter

# Calculate annual returns
df_annual_yahoo_stocks = df_yahoo_stocks.resample("Y").last()
df_annual_yahoo_stocks_rets = df_annual_yahoo_stocks / df_annual_yahoo_stocks.shift(1) -1
```

```python
# Plot annual returns as bar chart
df_annual_yahoo_stocks_rets.plot(kind="bar")
```
